```python
import pandas as pd
import gseapy as gp
```

```python
unique_genes_for_enrichment1 = pd.read_csv("unique_genes_for_enrichment1.csv")

gene_sets = [
    'KEGG_2021_Human',  
    'GO_Biological_Process_2021',  
    'Reactome_2022',  
    'MSigDB_Hallmark_2020'  
]


results = {}
for gene_set in gene_sets:
    enrichment = gp.enrichr(
        gene_list=unique_genes_for_enrichment1['Gene'].tolist(),
        gene_sets=gene_set,
        organism='Human',  
    )
    results[gene_set] = enrichment.results
```
